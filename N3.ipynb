```python

from matplotlib import pyplot as plt
from numpy import *
from itertools import combinations
from pandas import *
from seaborn import *
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# 1.3.1

ax, arr = plt.figure().add_subplot(111, projection = "3d"), random.rand(4, 3)

[ax.scatter(i[0], i[1], i[2]) for i in arr]

print("\tРасстояние Евклида\n", *[f"{linalg.norm(comb[0] - comb[1])}\n" for comb in combinations(arr, 2)])
print("\tКвадрат Евклидова расстояния\n", *[f"{linalg.norm((comb[0] - comb[1]) ** 2)}\n" for comb in combinations(arr, 2)])
print("\tРасстояние Чебыщева\n", *[f"{linalg.norm(comb[0] - comb[1], ord=inf)}\n" for comb in combinations(arr, 2)])
print("\tРасстояние Хемминга\n", *[f"{linalg.norm(comb[0] - comb[1], ord=1)}\n" for comb in combinations(arr, 2)])

```

```python
# 2.3.1

display(iris := load_dataset('iris'))
```

```python
# 2.3.1

plt.figure(figsize=(16, 7))
plt.subplot(121)

scatterplot(
    data = iris,
    x = "petal_length", y = "petal_width",
    hue = "species",
    s = 60,
)

plt.xlabel("Длинна лепестка, см")
plt.ylabel("Ширина лепестков, см")
plt.legend()
plt.grid()

plt.subplot(122)
scatterplot(
    data = iris,
    x = "sepal_length", y = "sepal_width",
    hue = "species",
    s = 60, 
)

plt.xlabel("Длинна чашелистика, см")
plt.ylabel("Ширина чашелистика, см")
plt.legend()
plt.grid()

```

```python
# 2.3.1

x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15
)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

plt.figure(figsize=(10, 7))
scatterplot(
    data = iris,
    x = "petal_length", y = "petal_width",
    hue = "species",
    s = 60,
)

plt.xlabel("Длинна лепестка, см")
plt.ylabel("Ширина лепестков, см")
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if array(y_test)[i] != y_pred[i]:
        plt.scatter(
            x_test.iloc[i, 2],
            x_test.iloc[i, 3],
            color = "red",
            s = 70 )

print(f'accuracy: {accuracy_score(y_test, y_pred):.3}')
```
